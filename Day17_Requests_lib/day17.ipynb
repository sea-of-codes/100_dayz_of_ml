```python
import requests
import pandas as pd
import time

api_key = "eaa9dapidad9633b4"
all_data = []

for i in range(1, 501):
    url = f"https://api.themoviedb.org/3/movie/popular?api_key={api_key}&language=en-US&page={i}"
    response = requests.get(url)
    data = response.json()

    if 'results' in data:
        df1 = pd.DataFrame(data['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        all_data.append(df1)
    else:
        print(f"API call failed for page {i}: {data.get('status_message', 'No status message')}")
        break
dff = pd.concat(all_data, ignore_index=True)

print(dff.shape)
print(dff.head())
```
